```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
pd.pandas.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('C:/Users/Raj/Documents/train.csv')
```

```python
train.head()
```

```python
categorical_features = [feature for feature in train.columns if train[feature].dtypes == 'O']
```

```python
####### Handling missing categorical features ####
cat_features_nan = [feature for feature in categorical_features if train[feature].isnull().sum()>1]
for feature in cat_features_nan:
    print("{}: {}% missing values".format(feature,np.round(train[feature].isnull().mean(),4)))
```

```python
def missing_cat_features(dataset,cat_features_nan):
  data = dataset.copy()
  data[cat_features_nan] = data[cat_features_nan].fillna('Missing')
  return data

train = missing_cat_features(train, cat_features_nan)

train[cat_features_nan].isnull().sum()
```

```python
#######Numerical variables ########
numerical_features = [feature for feature in train.columns if train[feature].dtypes != 'O']
```

```python
num_feature_nan = [feature for feature in numerical_features if train[feature].isnull().sum()>0]
num_feature_nan
```

```python
for feature in num_feature_nan:
  median_value = train[feature].median()
  train[feature+'nan'] = np.where(train[feature].isnull(),1,0)
  train[feature].fillna(median_value, inplace = True)

train[num_feature_nan].isnull().sum()
```

```python
train.head()
```

```python
###Handling temporal features #########
for feature in ['YearBuilt', 'YearRemodAdd','GarageYrBlt']:
  train[feature]=train['YrSold']-train[feature]
```

```python
train.head()
```

```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in num_features:
  train[feature] = np.log(train[feature])
```

```python
train.head()
```

```python
for feature in categorical_features:
    temp=train.groupby(feature)['SalePrice'].count()/len(train)
    temp_df=temp[temp>0.01].index
    train[feature]=np.where(train[feature].isin(temp_df),train[feature],'Rare_var')
```

```python
for feature in categorical_features:
    labels_ordered=train.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    train[feature]=train[feature].map(labels_ordered)
```

```python
train.head()
```

**Feature Scaling**

```python
####### Feature Scaling #######
features = [feature for feature in train.columns if feature not in ['Id', 'SalePrice']]

len(features)

```

```python
from sklearn.preprocessing import MinMaxScaler
scaler  = MinMaxScaler()
scaler.fit(train[features])
```

```python
scaler.transform(train[features])
```

```python
train_data = pd.concat([train[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(train[features]), columns=features)],
                    axis=1) 
```

```python
train_data.head()
```
