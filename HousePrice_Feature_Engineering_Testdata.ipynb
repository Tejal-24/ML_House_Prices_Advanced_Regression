```python
categorical_features = [feature for feature in test.columns if test[feature].dtypes == 'O']
```

```python
####### Handling missing categorical features ####
cat_features_nan = [feature for feature in categorical_features if test[feature].isnull().sum()>1]
for feature in cat_features_nan:
    print("{}: {}% missing values".format(feature,np.round(test[feature].isnull().mean(),4)))
```

```python
def missing_cat_features(dataset,cat_features_nan):
  data = dataset.copy()
  data[cat_features_nan] = data[cat_features_nan].fillna('Missing')
  return data

test = missing_cat_features(test, cat_features_nan)

test[cat_features_nan].isnull().sum()
```

```python
#######Numerical variables ########
numerical_features = [feature for feature in test.columns if test[feature].dtypes != 'O']
```

```python
num_feature_nan = [feature for feature in numerical_features if test[feature].isnull().sum()>0]
num_feature_nan
```

```python
for feature in num_feature_nan:
  median_value = test[feature].median()
  #test[feature+'nan'] = np.where(test[feature].isnull(),1,0)
  test[feature].fillna(median_value, inplace = True)

test[num_feature_nan].isnull().sum()
```

```python
test.head()
```

```python
test.shape
```

```python
###Handling temporal features #########
for feature in ['YearBuilt', 'YearRemodAdd','GarageYrBlt']:
  test[feature]=test['YrSold']-test[feature]
```

```python
test.head()
```

```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']
for feature in num_features:
  test[feature] = np.log(test[feature])
```

```python
test.head()
```

```python
for feature in categorical_features:
    temp=test.groupby(feature)['Id'].count()/len(test)
    temp_df=temp[temp>0.01].index
    test[feature]=np.where(test[feature].isin(temp_df),test[feature],'Rare_var')
```

```python
for feature in categorical_features:
    labels_ordered=test.groupby([feature])['Id'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    test[feature]=test[feature].map(labels_ordered)
```

```python
test.head()
```

**Feature Scaling**

```python
####### Feature Scaling #######
features = [feature for feature in test.columns if feature not in ['Id']]

len(features)

```

```python
from sklearn.preprocessing import MinMaxScaler
scaler  = MinMaxScaler()
scaler.fit(test[features])
```

```python
scaler.transform(test[features])
```

```python
test_data = pd.concat([test[['Id']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(test[features]), columns=features)],
                    axis=1) 
```
